```python
#finding columns with null values
movie_data.isnull().sum()
```

```python
# show rows with null value in Metascore column
row_null= movie_data[movie_data['Metascore'].isna()].index
row_null
```

```python
# show rows with null value in Gross column
row_null= movie_data[movie_data['Gross'].isna()].index
row_null
```

```python
#filling the missing values in Metascore with the median value
median= movie_data['Metascore'].median()
movie_data['Metascore']= movie_data['Metascore'].fillna(median)
movie_data['Metascore']
```

```python
#checking if Metascore as no missing values
movie_data.isna().sum()
```

```python
#removing the gross column
movie_data.drop(['Gross'], axis=1, inplace=True)
```

```python
#checking the data 
movie_data.head()
```

```python
#showing the values in the release year column
movie_data['Release Year'].unique()
```

```python
#cleaning the column
movie_data['Release Year']= movie_data['Release Year'].str.replace('Video', '')
movie_data['Release Year']= movie_data['Release Year'].str.replace('TV Movie', '')
movie_data['Release Year']= movie_data['Release Year'].str.replace('I', '')
movie_data['Release Year']= movie_data['Release Year'].str.replace('II', '')
```

```python
# result after cleaning
movie_data['Release Year'].unique()
```

```python
#Distribution of imdb rating
import pandas as pd
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
plt.hist(movie_data['IMDB Rating'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of IMDB Rating')
plt.xlabel('IMDB Rating')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
#IMDB Rating overtime
import pandas as pd
import matplotlib.pyplot as plt
avg_ratings_by_year= movie_data.groupby('Release Year')['IMDB Rating'].mean()
plt.figure(figsize=(10,6))
plt.plot(avg_ratings_by_year.index, avg_ratings_by_year.values, marker='o', color='green', linestyle='-')
plt.title('IMDB Rating')
plt.xlabel('Release Year')
plt.ylabel('Average imdb rating')
plt.grid(True)
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
#Average IMDB Ratings by genre 
import pandas as pd
import matplotlib.pyplot as plt
avg_ratings_by_genre= movie_data.groupby('Genre')['IMDB Rating'].mean()
plt.figure(figsize=(10,6))
avg_ratings_by_genre.plot(kind='hist', color='skyblue')
plt.title('Average IMDb Ratings by Genre')
plt.xlabel('Genre')
plt.ylabel('Average Imdb rating')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
#Correlation between imdb and metascore
import pandas as pd
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.scatter(movie_data['Metascore'], movie_data['IMDB Rating'], color='blue', alpha=0.5)
plt.title('IMDb rating vs Metascore ')
plt.xlabel('Metascore')
plt.ylabel('IMDb Rating')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Top rated movies
import pandas as pd
import matplotlib.pyplot as plt
top_rated_movies= movie_data.sort_values(by='IMDB Rating', ascending=False)
top_n= 10
top_rated_movies= top_rated_movies.head(top_n)
plt.figure(figsize=(10,6))
plt.bar(top_rated_movies['Movie Name'], top_rated_movies['IMDB Rating'], color='skyblue')
plt.title('Top Rated Movies')
plt.xlabel('Movie Name')
plt.ylabel('IMDB Rating')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
